```python
import pandas as pd
import numpy as np
data=pd.read_excel(".\data\\6.xlsx")
```

```python
data.head()
```

```python
bool_array = data.applymap(lambda x:type(x)!=int)
out_array = data[bool_array]
out_array.head()
```

```python
# 填充缺失值，填充为前一个或后一个（反正去重完只留为一个）
df=out_array.fillna(method='bfill',axis=1) 
df=out_array.fillna(method='ffill',axis=1) 

df.head()
```

```python
# df.to_excel("6.xlsx")
```

```python
def qc(x):
    a=list(np.sum(x))
    s=list(set(a))
    s.sort(key=a.index)
    for i in range(len(s)):
        s[i]=str(s[i])
    t="".join(s)
    return t
```
