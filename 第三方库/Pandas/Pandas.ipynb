#### Series
有点类似字典

```python
s1=pd.Series([1,"a",5,3,7])
```

```python
s1
# 左侧是索引，右边是数据
```

```python
# Series的索引不一定是数字，可以自己定义
s2=pd.Series([1,"a",5,2,7],index=["a","b","c","d","e"])
```

```python
s2
```

```python
# 利用字典创建Series
s={"a":1,"fsa":2,"fad":4}
s3=pd.Series(s)
s3
```

```python
# 访问Series中的数据
print(s3["a"])
```

DataFrame : 读进来的二位数据就是DataFrame

```python
fpath="./data/Boston.csv"
ratings=pd.read_csv(fpath)
```

```python
ratings.head()
```

```python
# 访问一列
ratings["zn"]
```

```python
# 访问多列
ratings[["zn","chas"]]
```

```python
# 访问一行
ratings.loc[1]
```

```python
# 访问多行
# 类似切片，但包含初值和末值
ratings.loc[1:3]
```

```python
# 访问一个数据
ratings.loc[1,"crim"]
```

```python
# 访问一小片
# 但这样不能用切片语法
ratings.loc[[1,3],["crim","zn"]]
```

```python
# 切片的话就不用两个中括号了
ratings.loc[1:3,"zn":"chas"]
```

#### 布尔访问


```python
df=ratings.loc[1:20,:]
df.shape
df.head(20)
```

```python
df.loc[df["zn"]<2,:]
# 可以看到6~12行被筛选掉了
```

```python
df.loc[(df["zn"]<2) & (df["indus"]<3),:]
# 十分强大啊
```

```python
df.loc[lambda df: (df["zn"]<2) & (df["indus"]<3),:]
# 甚至能用lambda表达式，给了更多可能
```
